### Importación de datos



```python
# Importación de librerías

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## I. Análisis Exploratorio de los datos EDA

### I.I Evaluacion de los datos

Como primer paso debemos de dar un primer vistazo a nuestros datos, con el propósito de detectar valores ausentes, erronéos, duplicados o con un formato no adecuado. Esto nos servirá para definir de mejor forma los pasos del procesamiento del dataset.

```python
# Estandarizamos el nombre de las columnas a un formato adecuado

def formato_columnas(df):
    df.columns = [col.lower().replace(' ', '_') for col in df.columns]
    return df.columns
```

```python
# Corrección del nombre de columnas
tienda.columns = formato_columnas(tienda)
tienda2.columns = formato_columnas(tienda2)
tienda3.columns = formato_columnas(tienda3)
tienda4.columns = formato_columnas(tienda4)

# Verificación 
print(tienda.columns)
print(tienda2.columns)
print(tienda3.columns)
print(tienda4.columns)
```

```python
tienda.head(10)
```

Los datos de todas las tiendas vienen limpios, de manera visual no se detecta la presencia de valores ausenten, sin embargo es conveniente verificarlo de manera aritmética.

Con el propósito de optimizar el procesado, podemos concatenar los datos de las tablas, con el fin de que solo sea un dataset

```python
tienda['tienda'] = 'Tienda 1'
tienda2['tienda'] = 'Tienda 2'
tienda3['tienda'] = 'Tienda 3'
tienda4['tienda'] = 'Tienda 4'



data = pd.concat([tienda, tienda2, tienda3, tienda4])
data.head()
```

```python
data.info()
```

# 1. Análisis de facturación

```python
# Ventas en cada tienda

ventas_por_tienda = (
    data
    .groupby('tienda')['precio']
    .sum()
)

display(ventas_por_tienda)
```

```python
# Gráfico de ventas por tienda

plt.figure(figsize = [8,5])
sns.set_style('ticks')
sns.barplot(data=ventas_por_tienda,
            width= 0.3,
            color = '#902b5c')
plt.xlabel('Tienda', fontsize = 12)
plt.ylabel('Ingresos totales ($ COP)', fontsize = 12)
plt.title('Ingresos totales por tienda', fontsize = 14)
plt.grid(axis='y', linestyle = '--')

plt.show()
```

El gráfico muestra que la tienda con mayor número de ventas es la Tienda 1, mientas que la Tienda 2 representa la tienda con menos ingresos totales.

# 2. Ventas por categoría

```python
# Ventas por categoría

ventas_categoria = (
    data
    .groupby(['tienda','categoría_del_producto'])
    .size()
    .reset_index(name = 'Cantidad de ventas')
    # .sort_values(by='Cantidad de ventas', ascending= False)
)

display(ventas_categoria)
```

```python
# Gráfico de ventas

plt.figure(figsize=[8,8])
sns.set_style('ticks')

sns.barplot(data=ventas_categoria,
            x= 'Cantidad de ventas',
            y= 'categoría_del_producto',
            hue= 'tienda',
            palette='rocket')

plt.title('Ventas por categorías', fontsize = 14, fontweight = 'bold')
plt.xlabel('Total de ventas', fontsize = 12)
plt.ylabel('Categoría', fontsize = 12)
plt.legend(title = 'Tienda')
plt.grid(axis='x', linestyle='--')

plt.show()
```

Se aprecia que en todas las tiendas la categoría de productos más vendida es Muebles, seguido por Electrónicos.

# 3. Calificación promedio de la tienda


```python
# Calfiicacion de los usuarios por tienda

calificacion = (
    data
    .groupby('tienda')['calificación']
    .mean()
)

display(calificacion.sort_values(ascending=False))
```

Si bien la tiendas tienen una calificación muy similar, que solo varía por decimales, la tienda con calificación más baja es la tienda 1, siendo la tienda 3 la mejor calificada por los usuarios.

# 4. Productos más y menos vendidos

```python
# Productos más vendidos por tienda

most_sold = (
    data.groupby(['tienda', 'producto'])['producto']
    .count()
    .reset_index(name= 'total_vendidos')
)
```

```python
display(most_sold)
```

```python
# Separamos por tiendas, seleccionando los 3 productos mas vendidos y 3 productos menos vendidos

# Para la tienda 1

productos_tienda_1 = (
    most_sold[most_sold['tienda'] == 'Tienda 1']
    .sort_values(by = 'total_vendidos', ascending=False)
    .reset_index(drop=True)
)

# Para la tienda 2

productos_tienda_2 = (
    most_sold[most_sold['tienda'] == 'Tienda 2']
    .sort_values(by = 'total_vendidos', ascending=False)
    .reset_index(drop=True)
)

# Para la tienda 3

productos_tienda_3 = (
    most_sold[most_sold['tienda'] == 'Tienda 3']
    .sort_values(by = 'total_vendidos', ascending=False)
    .reset_index(drop=True)
)

# Para la tienda 4

productos_tienda_4 = (
    most_sold[most_sold['tienda'] == 'Tienda 4']
    .sort_values(by = 'total_vendidos', ascending=False)
    .reset_index(drop=True)
)
```

```python
# Concatenamos los 3 primeros y los 3 últimos
productos_tienda_1 = pd.concat([productos_tienda_1.head(3), productos_tienda_1.tail(3)])
productos_tienda_2 = pd.concat([productos_tienda_2.head(3), productos_tienda_2.tail(3)])
productos_tienda_3 = pd.concat([productos_tienda_3.head(3), productos_tienda_3.tail(3)])
productos_tienda_4 = pd.concat([productos_tienda_4.head(3), productos_tienda_4.tail(3)])
```

```python
print(productos_tienda_1)
```

```python
# lista de dataFrame con los productos mas vendidos por tienda

most_sold_df = [productos_tienda_1, productos_tienda_2, productos_tienda_3, productos_tienda_4]

fig, axs = plt.subplots(2,2, figsize = [14, 8])
axs = axs.flatten()

for i, df in enumerate(most_sold_df):
    sns.set_theme()
    sns.barplot(data= df, 
                x= 'total_vendidos', 
                y = 'producto', 
                ax=axs[i],
                width= 0.3, 
                color='#5975a4')
    axs[i].tick_params(axis='x', rotation=0)
    axs[i].grid(axis='x', linestyle='--', color='gray', alpha=0.6)
    axs[i].set_title(f'Ventas Tienda {i+1}')
    axs[i].set_xlabel('Producto')
    axs[i].set_ylabel('Total vendido')

fig.suptitle('Análisis Comparativo de Ventas: Todas las Tiendas', 
            fontsize=22, 
            fontweight='bold'
)
plt.tight_layout()



plt.show()
```

La información de productos más vendidos corrobora la información mostrada en el apartado 2, vemos que los productos más vendidos corresponden a muebles. 

# 5. Envío promedio por tienda

```python
# Agrupamos para cada tienda

costo_de_envío = round(
    data
    .groupby('tienda')['costo_de_envío']
    .mean()
    ,2                  # Redondeo de decimales del costo de envío
)
```

```python
costo_de_envío.plot(kind = 'bar',
                    width = 0.3)
plt.title('Costo de envío promedio', 
            fontsize = 14, 
            weight = 'bold')
plt.ylabel('Pesos ($COP)', fontsize = 12)
plt.xlabel('Tienda', fontsize = 12)
plt.xticks(rotation = 0)

plt.show()
```
